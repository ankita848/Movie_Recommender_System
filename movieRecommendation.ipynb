```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('tmdb_5000_movies.csv')
data2 = pd.read_csv('tmdb_5000_credits.csv')
```

```python
data = data1.merge(data2, on = 'title')
```

```python
data.head(1)
```

```python
data.info()
```

```python
data = data[['id', 'keywords', 'title', 'genres', 'overview', 'cast', 'crew']]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.iloc[0].genres
```

```python
import ast
def convert_list(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
data['genres'] = data['genres'].apply(convert_list)
```

```python
data['keywords'] = data['keywords'].apply(convert_list)
```

```python
data.head()
```

```python
import ast
def convert3(obj):
    l = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count += 1
        else:
            break;
    return l
```

```python
data['cast'] = data['cast'].apply(convert3)
```

```python
data.head()
```

```python
def fetch_dir(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
data['crew'] = data['crew'].apply(fetch_dir)
```

```python
data['overview']
```

```python
data.head()
```

```python
data['overview'] = data['overview'].apply(lambda x: x.split())
```

```python
data.head(2)
```

```python
data['keywords'] = data['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
data['genres'] = data['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
data['cast'] = data['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
data['crew'] = data['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
data.head(2)
```

```python
data['tags'] = data['keywords'] + data['genres'] + data['overview'] + data['cast'] + data['crew']
```

```python
df = data[['id', 'title', 'tags']]
```

```python
df['tags']
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags']
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(obj):
    y = []
    for i in obj.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    idx = df[df['title'] == movie].index[0]
    distances = similarity[idx]
    list_movie = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1: 6]
    for i in list_movie:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
